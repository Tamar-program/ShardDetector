```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt

#Loading the model
model = YOLO("yolov8n.pt")
#Running the model on a random image
results = model(r"C:\Users\user\Desktop\iaa-pictures\111.png")  

#Showing an image with identical YOLO objects
plt.figure(figsize=(8,6))
plt.imshow(results[0].plot())  # plot() Returns the image with the boxes drawn.
plt.axis('off')
plt.title("The picture with the boxes")
plt.show()
```
